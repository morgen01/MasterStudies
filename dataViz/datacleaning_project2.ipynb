```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gdp = pd.read_csv('gdp.csv', encoding = "ISO-8859-1", header=1)
gdp.head()
```

```python
gdp = gdp.drop(gdp.filter(regex='Unnamed').columns, axis=1)
gdp.head()
```

```python
gdp.columns
```

```python
gdp.tail()
```

```python
gdp.dropna(axis=0, how='any', inplace=True)
gdp.tail()
```

```python
gdp = gdp[gdp['1990']!= '..']
```

```python
gdp = gdp[['HDI Rank (2017)', 'Country', '1995', '2000', '2005', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']]
gdp.head()
```

```python
# merge all 3 dataframes
frames = [gdp, unemployment, life_exp]
df_keys = pd.concat(frames, keys=['gdp', 'unemployment', 'life_exp'])
df_keys
```

```python
df_keys.sort_values(by=['HDI Rank (2017)'])
```
